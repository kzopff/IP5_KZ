Using the corr() fucntion, I made a correlation matrix with the quantitative data from the Automobile Data Set.

```python
df.corr()
```

```python
corrdf = df.corr()
```

Then, I used Seaborn to create a heatmap of the correlation data.

```python
sns.heatmap(corrdf, annot=True)
plt.savefig('corr_heatmap.png')
```

Creating a figure, I added the histograms of all of the quantitative data to one image.

```python
fig = plt.figure(figsize=(15,22))
ax1 = fig.add_subplot(5,2,1)
ax1.hist(df.symboling)
ax1.set_title('Symboling')
ax2 = fig.add_subplot(5,2,2)
ax2.hist(df.wheel_base)
ax2.set_title('Wheel Base')
ax3 = fig.add_subplot(5,2,3)
ax3.hist(df.length)
ax3.set_title('Length')
ax4 = fig.add_subplot(5,2,4)
ax4.hist(df.width)
ax4.set_title('Width')
ax5 = fig.add_subplot(5,2,5)
ax5.hist(df.height)
ax5.set_title('Height')
ax6 = fig.add_subplot(5,2,6)
ax6.hist(df.curb_weight)
ax6.set_title('Curb Weight')
ax7 = fig.add_subplot(5,2,7)
ax7.hist(df.engine_size)
ax7.set_title('Engine Size')
ax8 = fig.add_subplot(5,2,8)
ax8.hist(df.compression_ratio)
ax8.set_title('Compression Ratio')
ax9 = fig.add_subplot(5,2,9)
ax9.hist(df.city_mpg)
ax9.set_title('City MPG')
ax10 = fig.add_subplot(5,2,10)
ax10.hist(df.highway_mpg)
ax10.set_title('Highway MPG')
plt.savefig('densityplts.png')
```

Then, I used Seaborn to create a pair plot of all of the quantitative data.

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(df, hue='body_style')
plt.savefig('pairplot.png')
```

creating another figure, I added horizontal bar charts for each of the categorical variables to one image.

```python
fig = plt.figure(figsize=(15,22))
ax1 = fig.add_subplot(5,2,1)
ax1 = make.sort_values().plot.barh()
ax1.set_title('Make')
ax2 = fig.add_subplot(5,2,2)
ax2 = fuel_type.sort_values().plot.barh(color='red')
ax2.set_title('Fuel Type')
ax3 = fig.add_subplot(5,2,3)
ax3 = aspiration.sort_values().plot.barh(color='green')
ax3.set_title('Aspiration')
ax4 = fig.add_subplot(5,2,4)
ax4 = num_of_doors.sort_values().plot.barh(color='purple')
ax4.set_title('Number of Doors')
ax5 = fig.add_subplot(5,2,5)
ax5 = body_style.sort_values().plot.barh()
ax5.set_title('Body Style')
ax6 = fig.add_subplot(5,2,6)
ax6 = drive_wheels.sort_values().plot.barh(color='red')
ax6.set_title('Drive Wheels')
ax7 = fig.add_subplot(5,2,7)
ax7 = engine_location.sort_values().plot.barh(color='green')
ax7.set_title('Engine Location')
ax8 = fig.add_subplot(5,2,8)
ax8 = engine_type.sort_values().plot.barh(color='purple')
ax8.set_title('Engine Type')
ax9 = fig.add_subplot(5,2,9)
ax9 = num_of_cylinders.sort_values().plot.barh()
ax9.set_title('Number of Cylinders')
ax10 = fig.add_subplot(5,2,10)
ax10= fuel_system.sort_values().plot.barh(color='red')
ax10.set_title('Fuel System')
plt.savefig('barcharts.png')
```

Here, I used a scatter plot and a regression model for the comparison of the city_mpg and engine_size variables to better explore their negative correlation and others like it. 

```python
sns.scatterplot(x=df['city_mpg'], y=df['engine_size'])
plt.savefig('city_mpg_engine_size_scatter.png')
```

```python
sns.regplot(x=df['city_mpg'], y=df['engine_size'], data=df)
plt.savefig('city_mpg_engine_size_regression.png')
```
